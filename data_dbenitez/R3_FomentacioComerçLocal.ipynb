```python
# Crear un nuevo DataFrame con las columnas seleccionadas
df_json = dataset_cens_plantaBaixa[["Nom_Local", "Nom_Activitat2", "Nom_Barri", "Nom_Districte", "Latitud", "Longitud"]]

# Convertir el DataFrame a formato JSON
json_data = df_json.to_json(orient="records")

# Imprimir el JSON resultante
print(json_data)
```

```python
# Nombre del archivo y ruta de la carpeta donde se guardará el JSON
carpeta = "json"
nombre_archivo = "cens_datos.json"
ruta_archivo = os.path.join(carpeta, nombre_archivo)

# Crear la carpeta si no existe
os.makedirs(carpeta, exist_ok=True)

# Guardar el JSON en el archivo
with open(ruta_archivo, 'w') as archivo:
    json.dump(json_data, archivo)

print(f"JSON guardado en {ruta_archivo}")
```
